```python
import cv2
import numpy as np
```

```python
img=cv2.imread("/Users/lakshmikeerthanaaravapalli/Desktop/ComputerVision/col_balls.jpg",0)
```

```python
#erosion, dilation

_,img_mask=cv2.threshold(img,230,255,cv2.THRESH_BINARY_INV)
kernel=np.ones((5,5),np.uint8)

#works on object edges
e=cv2.erode(img_mask,kernel)

kernel2=np.ones((3,3),np.uint8)

#works on distortions present in object
d=cv2.dilate(img_mask,kernel2)
```

```python
#cv2.imshow("eroded",e)
#cv2.imshow("dilated",d)

from matplotlib import pyplot as plt
titles = ["img","mask","erosion","dilation"]
images = [img,img_mask,e,d]
for i in range(4):
    plt.subplot(2, 2, i+1), plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])

plt.show()


#cv2.waitKey(0)
#cv2.destroyAllWindows()
```

```python
#opening,closing

o=cv2.morphologyEx(img_mask,cv2.MORPH_OPEN,kernel2)

c=cv2.morphologyEx(img_mask,cv2.MORPH_CLOSE,kernel2)
```

```python
from matplotlib import pyplot as plt
titles = ["opening","closing"]
images = [o,c]
for i in range(len(titles)):
    plt.subplot(2, 2, i+1), plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])

plt.show()
```
