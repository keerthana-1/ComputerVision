```python
import cv2
import numpy as np
```

```python
#load image into gray scale
img = cv2.imread("/Users/lakshmikeerthanaaravapalli/Desktop/ComputerVision/building.jpg")
img = cv2.resize(img,(600,600))
img_gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
def nothing(x):
    pass

cv2.namedWindow("Canny")
threshold=cv2.createTrackbar("Threshold","Canny",0,255,nothing)

while True:
    a=cv2.getTrackbarPos("Threshold","Canny")
    #Canny(img,thresh1,thresh2) 
    canny=cv2.Canny(img_gray,a,255)
    cv2.imshow("Canny",canny)
    k=cv2.waitKey(1)&0xFF
    if(k==27):
        break
    
cv2.destroyAllWindows()


```

```python
titles = ["original","Canny"]
images = [img,canny]

#if you want then plot it
from matplotlib import pyplot as plt
for i in range(len(titles)):
    plt.subplot(1,2, i+1), 
    plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
```
